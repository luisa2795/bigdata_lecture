```python
#transformation to Tidy Data Format: 5 columns (country, year and all three KPIs from the three dataframes)
#tidy_data=pd.Dataframe()

life_tidy=pd.melt(life_exp, id_vars=['country'], var_name='year', value_name='life_expectancy')
life_tidy
```

```python
pop_tidy=pd.melt(pop, id_vars=['country'], var_name='year', value_name='population')
gni_tidy=pd.melt(gni_pc, id_vars=['country'], var_name='year', value_name='GNI_per_cap')
gni_tidy
```

```python
merge1=life_tidy.merge(pop_tidy, how='outer', on=['country', 'year'])
```

```python
tidy_df=merge1.merge(gni_tidy, how='outer', on=['country', 'year'])
```

```python
tidy_df['life_expectancy']=pd.to_numeric(tidy_df['life_expectancy'])
```
